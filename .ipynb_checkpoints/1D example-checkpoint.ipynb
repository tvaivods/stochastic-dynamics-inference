# 1D example from the stochastic SINDy paper

Want to infer dynamics
$$
\dot{x} = -U'(x) = 3 - 18x + 12x^2 - 2x^3.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from time_series import *
from stochastic_sindy import *

y = lambda x : 3 - 18*x  + 12*x**2 - 2*x**3
x = np.linspace(-1,5,100)
plt.plot(x,y(x))
plt.show()
```

```python
xs = time_series(y, -1, 0.01, 100, 0.1)

plt.plot(x,y(x))
plt.scatter(xs,y(xs), color = 'red')
plt.show()
```

```python
x_multiple = []
dt = 0.0001
n = 10000
n_traj = 10

for x0 in np.linspace(-1,5,n_traj):
    x_multiple.append(time_series(y, x0, dt, n, 1.2))

for i in range(n_traj):
    plt.plot(x_multiple[i],np.arange(0,(n+1)*dt,dt))
plt.xlabel("x")
plt.ylabel("t")
plt.show()
```

```python
y_multiple = ps.differentiation.FiniteDifference()._differentiate(np.array(x_multiple).T,dt)

# Stack the data into one column
y_single = y_multiple.reshape(-1,1)
x_single = np.array(x_multiple).reshape(-1,1)

bin_n = 20
x_binned, y_binned, weights = bin_data(x_single, y_single, bin_n)

basis = ps.feature_library.polynomial_library.PolynomialLibrary(degree=3)
basis.fit(x_binned)
X = basis.transform(x_binned)
W = np.diag(weights)
```

```python
errors = CV_score(X,Y,K=4)
plt.plot(np.arange(X.shape[1] * Y.T.shape[0], 0, -1), errors, '-o')
plt.xlabel("Total number of terms (sparsity)")
plt.ylabel("Cross-validation score $\delta$")
plt.show()
```

```python
mask = None
N = X.shape[1] * Y.shape[1]
for i in range(N):
    C, mask = SSR(X, Y, mask)
    print(C)

model = ps.SINDy(feature_names = "x", feature_library = ps.PolynomialLibrary(degree = 3))
model.fit(xs, t=dt, multiple_trajectories = True)   

model.print()
```
